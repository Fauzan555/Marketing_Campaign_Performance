# Data Cleaning & Preprocessing

```python
df.isnull().sum()
```

```python
df.duplicated().sum()
```

```python
df = df.drop_duplicates()
```

# Date Formatting

```python
df['Date'] = pd.to_datetime(df['Date'])
df['Month'] = df['Date'].dt.to_period('M').astype(str)
```

# Feature Engineering

# TR (Click Through Rate)

```python
df['CTR'] = df['Clicks'] / df['Impressions']
```

```python
df['Acquisition_Cost'] = pd.to_numeric(df['Acquisition_Cost'], errors='coerce')
```

# ROI per Cost Unit

```python
df['ROI_per_Cost'] = df['ROI'] / df['Acquisition_Cost']
df
```

# Overall Campaign Performance

```python
df[['Conversion_Rate','ROI','Acquisition_Cost','CTR','Engagement_Score']].mean()
```

```python
df['Acquisition_Cost'].isna().sum(), len(df)
```

```python
df['Acquisition_Cost'].dtype
```

```python
df['Acquisition_Cost'] = pd.to_numeric(df['Acquisition_Cost'], errors='coerce')
```

```python
df.groupby('Channel_Used')['Acquisition_Cost'].mean()
```

```python
df['Acquisition_Cost'] = df.groupby('Channel_Used')['Acquisition_Cost']\
    .transform(lambda x: x.fillna(x.median()))
```

```python
df['Acquisition_Cost'] = df['Acquisition_Cost'].fillna(
    df['Acquisition_Cost'].median()
)
```

```python
df[['Acquisition_Cost']].describe()
```

```python
df['ROI_per_Cost'] = df['ROI'] / df['Acquisition_Cost']
```

```python
df['Cost_per_Conversion'] = df['Acquisition_Cost'] / df['Conversion_Rate']
```

# Channel Performance Analysis

```python
channel_cost = df.groupby('Channel_Used')[[
    'ROI',
    'Acquisition_Cost',
    'ROI_per_Cost'
]].mean().sort_values('ROI_per_Cost', ascending=False)

channel_cost
```

```python
df.groupby('Channel_Used')['ROI'].mean()
```

```python
df['ROI_per_Engagement'] = df['ROI'] / df['Engagement_Score']
```

```python
df['ROI_per_CTR'] = df['ROI'] / df['CTR']
```

```python
# Count the occurrences of each channel used
channel_counts = df['Channel_Used'].value_counts()

# Create a pie chart to visualize the distribution of channel usage
plt.figure(figsize=(8, 8))
plt.pie(channel_counts, labels=channel_counts.index, autopct='%1.1f%%')
plt.title('Distribution of Channel Usage')
plt.show()
```

# Campaign Type vs ROI

```python
sns.boxplot(data=df, x='Campaign_Type', y='ROI')
plt.xticks(rotation=45)
plt.title('ROI Distribution by Campaign Type')
plt.show()
```

# Funnel Analysis (Impression → Click → Conversion)

```python
funnel = df[['Impressions','Clicks','Conversion_Rate']].mean()
funnel
```

```python
# Scatter plot of Clicks vs. Impressions
plt.figure(figsize=(10, 6))
sns.scatterplot(data=df, x='Impressions', y='Clicks', hue='Campaign_Type')
plt.title('Scatter plot of Clicks vs. Impressions')
plt.xlabel('Impressions')
plt.ylabel('Clicks')
plt.show()
```

```python
sns.barplot(x=funnel.index, y=funnel.values)
plt.title('Average Marketing Funnel')
plt.show()
```

# Customer Segment Analysis

```python
segment_perf = df.groupby('Customer_Segment')[[
    'Conversion_Rate','ROI','Engagement_Score'
]].mean().sort_values('ROI', ascending=False)

segment_perf
```

```python
segment_perf.plot(kind='bar', figsize=(10,5))
plt.title('Customer Segment Performance')
plt.show()
```

# Target Audience & Localization Analysis

```python
# Target Audience
df.groupby('Target_Audience')['ROI'].mean().sort_values(ascending=False)
```

```python
# Language

df.groupby('Language')['Conversion_Rate'].mean().sort_values(ascending=False)
```

# Engagement Threshold Discovery

```python
df['High_Conversion'] = df['Conversion_Rate'] > df['Conversion_Rate'].median()

df.groupby('High_Conversion')['Engagement_Score'].mean()
```

```python
# Histogram of Conversion Rate
plt.figure(figsize=(10, 6))
sns.histplot(data=df, x='Conversion_Rate', bins=20)
plt.title('Histogram of Conversion Rate')
plt.xlabel('Conversion Rate')
plt.ylabel('Count')
plt.show()
```
